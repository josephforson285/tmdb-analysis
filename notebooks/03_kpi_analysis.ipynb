## 3.2 Advanced search queries

Because we flattened genres and cast into `|`-separated strings, filtering is
vectorized string matching: **`.str.contains()`** gives a boolean mask per row,
and `&` combines masks. `na=False` says "a missing value is not a match"
(without it, NaN cells would poison the mask).

>  Our 18
> ids are all modern blockbusters — none stars Bruce Willis or Uma Thurman. The
> point is that the *query logic* is correct and reusable on any larger dataset.

```python
# Search 1: Sci-Fi Action movies starring Bruce Willis, best-rated first
search_1 = movies[
    movies["genres"].str.contains("Science Fiction", na=False)
    & movies["genres"].str.contains("Action", na=False)
    & movies["cast"].str.contains("Bruce Willis", na=False)
].sort_values("vote_average", ascending=False)

print(f"{len(search_1)} matches")
search_1[["title", "genres", "vote_average"]]
```

```python
# Search 2: Uma Thurman movies directed by Quentin Tarantino, shortest first
# (.str.contains rather than == because 'director' can hold co-directors: 'A|B')
search_2 = movies[
    movies["cast"].str.contains("Uma Thurman", na=False)
    & movies["director"].str.contains("Quentin Tarantino", na=False)
].sort_values("runtime", ascending=True)

print(f"{len(search_2)} matches")
search_2[["title", "director", "runtime"]]
```

```python
# Proof the pattern works — same query shape, people who ARE in our data:
check = movies[
    movies["genres"].str.contains("Action", na=False)
    & movies["cast"].str.contains("Robert Downey Jr.", na=False)
].sort_values("vote_average", ascending=False)
check[["title", "vote_average"]]
```

## 3.3 Franchise vs. standalone performance

This is done with **`groupby`**. The recipe is
*split → apply → combine*: split rows into groups, apply an aggregation to each,
combine into a summary table.

<!-- First we need a grouping key: franchise membership is simply "does
`belongs_to_collection` have a value?" -->

```python
movies["franchise"] = np.where(movies["belongs_to_collection"].notna(), "Franchise", "Standalone")
movies["franchise"].value_counts()
```

```python
franchise_vs_standalone = movies.groupby("franchise").agg(
    mean_revenue=("revenue_musd", "mean"),
    median_roi=("roi", "median"),
    mean_budget=("budget_musd", "mean"),
    mean_popularity=("popularity", "mean"),
    mean_rating=("vote_average", "mean"),
).round(2)

franchise_vs_standalone
```

## 3.4 Most successful franchises

```python
top_franchises = (
    movies.dropna(subset=["belongs_to_collection"])
    .groupby("belongs_to_collection")
    .agg(
        num_movies=("id", "count"),
        total_budget=("budget_musd", "sum"),
        mean_budget=("budget_musd", "mean"),
        total_revenue=("revenue_musd", "sum"),
        mean_revenue=("revenue_musd", "mean"),
        mean_rating=("vote_average", "mean"),
    )
    .sort_values("total_revenue", ascending=False)
    .round(2)
)
top_franchises
```

## 3.5 Most successful directors

 Four of our movies are co-directed  both
Avengers films (Joe & Anthony Russo) and both Frozen films (Chris Buck &
Jennifer Lee), so `director` holds e.g. `'Joe Russo|Anthony Russo'`. A naive
`groupby("director")` would treat every distinct string as a different person
and split a duo's track record across half-empty rows.

The fix is **`.explode()`**: after `.str.split("|")` each cell holds a *list* of
directors; explode gives every list element its own row (duplicating the movie's
other columns). Each person then gets full credit for each movie they directed,
and `groupby` counts people, not strings.

```python
top_directors = (
    movies.assign(director=movies["director"].str.split("|"))
    .explode("director")
    .groupby("director")
    .agg(
        num_movies=("id", "count"),
        total_revenue=("revenue_musd", "sum"),
        mean_rating=("vote_average", "mean"),
    )
    .sort_values("total_revenue", ascending=False)
    .round(2)
)
top_directors.head(10)
```
